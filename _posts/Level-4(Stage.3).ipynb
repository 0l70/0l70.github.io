### **for 반복문** : 범위와 함께 사용하기

> for 숫자 변수 in 범위:<br>&#160;&#160;&#160;&#160;&#160;&#160;코드

```python
#for 반복문과 범위를 함께 조합해서 사용
for i in range(5):
    print(str(i) + " = 반복 변수")
print()

for i in range(5, 10):
    print(str(i) + " = 반복 변수")
print()

for i in range(0, 10, 3):
    print(str(i) + " = 반복 변수")
print()
```

### **for 반복문** : 리스트와 함께 조합하기

```python
# 리스트 선언
array = [273, 32, 103, 57 ,52]

# 리스트에 반복문 적용
for element in array:
    print(element)
```

### 리스트와 범위를 조합해서 사용하기

```python
array = [273, 32, 103, 57 ,52]

for i in range(len(array)):
    print("{}번째 반복 : {}".format(i, array[i]))
```

### 중첩 반복문으로 피라미드 만들기

- 피라미드는 인간이 만든 것이 맞을까?

```python
output = ""

for i in range(1,10):
    for j in range(0,i):
        output += "*"
    output += "\n"

print(output)
```

```python
output = ""

for i in range(1,15):
    for j in range(14,i,-1):
        output += " "
    for k in range(0, 2*i-1):
        output += "*"
    output += "\n"

print(output)
```

### 크리스마스 트리 만들기(좀 따 해야지 ㅋ)

```python
output = ""

for i in range(1,15):
    for j in range(14,i,-1):
        output += " "
    for k in range(0, 2*i-1):
        output += "*"
    output += "\n"

print(output)

```

### while 반복문

> while 불 표현식: <br> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 문장


```python
# while 반복문 사용
while True:
    # '.'을 출력
    # 기본적으로 end가 '\n'이라 줄바꿈이 일어나는데
    # 빈 문자열 ""로 바꿔서 줄바꿈이 일어나지 않게 한다.
    print(".", end = "")
```

### **while 반복문** : for 반복문처럼 사용하기

```python
# 반복 변수를 기반으로 반복
i = 0
while i < 10:
    print("{}번째 반복입니다.", format(i))
    i += 1
```

### **while 반복문** : 상태를 기반으로 반복하기

- remove(value)

```python
# 변수 선언
list_test = [1, 2, 1, 2]
value = 2

# list_test 내부에 value가 있다면 반복
while value in list_test:
    list_test.remove(value)

# 출력
print(list_test)
```

### **while 반복문** : 시간을 기반으로 반복

#### 유닉스 타임(Unix Time) 기준

```python
import time

time.time()
```

```python
# 시간과 관련된 기능을 가져온다.
import time

#변수를 선언
number = 0

# 5초 동안 반복
target_tick = time.time() + 5
while time.time() < target_tick:
    number += 1

print('5초 동안 {}번 반복했습니다.'.format(number))
```

### **while 반복문** : break 키워드 / continue 키워드

#### **break 키워드**

> 반복문을 벗어나는 키워드

```python
# 변수 선언
i = 0

# 무한 반복
while True:
    # 몇 번째 반복
    print("{}번째 반복문입니다.".format(i))
    i= i+1

    # 반복 종료
    input_text = input("> 종료하시겠습니까? (y/n) :")
    if input_text in ["y", "Y"]:
        print("반복은 종료합니다.")
        break
```

#### **continue 키워드**

- 현재 반복 생략 후 다음 반복으로 넘어갈 때 사용하는 키워드

```python
# 변수 선언
numbers = [5, 15, 6, 20, 7, 25]

# 반복 돌리기
for number in numbers:
    # number가 10보다 작으면 다음 반복으로 넘어감
    if number < 10:
        continue
    # 출력
    print(number)
```
